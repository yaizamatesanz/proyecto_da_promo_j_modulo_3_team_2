### VALORES ÚNICOS

Lo hemos hecho de dos formas:
1) Mediante un bucle for iterando por las columnas categoricas
2) Seleccionando columna a columna los valores unicos

```python
# Sacamos los valores únicos y su frecuencia:
for columna in columnas_categoricas:
    print(f"{columna.upper()}'\n")
    print(f"Valores únicos de las columnas categoricas: {df_categoricas[columna].unique()}\n")
    print(f"Frecuencia de los valores únicos: {df_categoricas[columna].value_counts()}")
```

### NULOS

```python
# Verificamos los valores nulos por cada columna:
df_hr.isnull().sum()
```

```python
# Verificamos los valores no nulos por cada columna:
df_hr.notnull().sum()
```

```python
# Calculamos el porcentaje de nulos que tenemos para cada columna
nulos = (df_hr.isnull().sum() / df_hr.shape[0]) * 100
nulos
```

```python
# Lo convertimos a DataFrame
df_hr_nulos = pd.DataFrame(nulos, columns = ["%_nulos"])

# Filtramos el DataFrame para quedarnos solo con aquellas columnas que tengan nulos
df_hr_nulos[df_hr_nulos["%_nulos"] > 0]
```
